```python
import time

import matplotlib.pyplot as plt

from fibonacci import *
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('fib_results.csv')
df[df["mode"] == "iter"][["n","time_ns"]].groupby("n").mean()

# Group by mode and n, compute mean time
grouped = df.groupby(["mode", "n"])["time_ns"].mean().unstack("mode")

# Plot all modes on the same chart
grouped.plot()
plt.title("Tempo Médio x n x Algoritmo")
plt.xlabel("n")
plt.ylabel("Tempo Médio (ns)")
plt.grid(True)
plt.legend(title="Algoritmo")
plt.tight_layout()
plt.show()
```

```python
df_pol = pd.read_csv('pol_results.csv')
# Group by mode and n, compute mean time
grouped_pol = df_pol.groupby(["mode", "n"])["time_ns"].mean().unstack("mode")

# Plot all modes on the same chart
grouped_pol.plot()
plt.title("Tempo Médio x n x Algoritmo")
plt.xlabel("n")
plt.ylabel("Tempo Médio (ns)")
plt.grid(True)
plt.legend(title="Algoritmo")
plt.tight_layout()
plt.show()
```

# Complexidade Teórica

### Fibonacci
- Recursivo
    - Com IA
        - T(0) = 1
        - T(2) = 1
        - T(n) = T(n-1) + T(n-2)
        - O(2^n)
    - Sem IA
        - T(0) = 1
        - T(2) = 1
        - T(n) = T(n-1) + T(n-2)
        - O(2^n)
- Iterativo
    - Com IA
        - T(n) = 4n + 4
        - O(n)
    - Sem IA
        - T(n) = 4n + 5
        - O(n)



### Avaliação de Polinômios (Horner)
- Iterativo
    - Com IA
        - 2n
        - O(n)
    - Sem IA
        - 2n
        - O(n)
- Recursivo
    - Com IA
        - T(0) = 1
        - T(n) = T(n-1) + 1
        - O(2^n)
    - Sem IA
        - T(0) = 1
        - T(n) = T(n-1) + 1
        - O(2^n)
