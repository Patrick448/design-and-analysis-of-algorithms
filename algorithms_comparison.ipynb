```python
import time

import matplotlib.pyplot as plt

from fibonacci import *
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('fib_results.csv')
df[df["mode"] == "iter"][["n","time_ns"]].groupby("n").mean()
```

```python


df[df["mode"] == "iter"][["n","time_ns"]].groupby("n").mean().plot()
df[df["mode"] == "rec"][["n","time_ns"]].groupby("n").mean().plot()
```

```python
# Group by mode and n, compute mean time
grouped = df.groupby(["mode", "n"])["time_ns"].mean().unstack("mode")

# Plot all modes on the same chart
grouped.plot(marker='o')
plt.title("Average Time vs n by Mode")
plt.xlabel("n")
plt.ylabel("Average Time (ns)")
plt.grid(True)
plt.legend(title="Mode")
plt.tight_layout()
plt.show()
```
